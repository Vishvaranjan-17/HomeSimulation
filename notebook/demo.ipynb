# Voice-Based Smart Home Simulator - Demo

End-to-end walkthrough of the fallback pipeline.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
from smarthome.pipeline import Pipeline
p = Pipeline()
print('Devices:', p.store.device_names())
```

## Basic commands

```python
for cmd in ['Turn on bedroom light', 'Set AC to 24 degrees', "What's the AC temperature?"]:
    r = p.handle(cmd)
    print(cmd, '->', r.intent, r.slots, '|', r.response)
```

## Multi-turn context + relative command

```python
print(p.handle('Turn on the TV').response)
print(p.handle('Turn it off').response)  # resolved from context
print(p.handle('Set AC to 24 degrees').response)
print(p.handle('increase AC by 2 degrees').response)
```

## Ambiguous + routine + Hindi

```python
p2 = Pipeline()
print(p2.handle('Turn it off').response)  # asks to clarify
print(p2.handle('Goodnight').response)
print(p2.handle('bedroom light band karo').response)
```

## Accuracy analysis over the test set

```python
import json
cases = json.load(open('../data/test_commands.json'))['commands']
correct = 0
for c in cases:
    pp = Pipeline()
    if c.get('ambiguous'):
        pp.handle('Turn on the AC')
    out = pp.handle(c['text'])
    correct += int(out.intent == c['intent'])
print(f'Intent accuracy: {correct}/{len(cases)} = {correct/len(cases):.1%}')
```
